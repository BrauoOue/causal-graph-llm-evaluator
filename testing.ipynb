```python
class Mapping:

    def __init__(self, context, question, question_type, choices, label, explanation):
        self.label = label
        self.choices = choices
        self.question_type = question_type
        self.question = question
        self.context = context
        self.explanation = explanation

    def get_columns(self):
        return ["context", "question", "question_type", "label", "explanation"]


```

MainCLass, Builder/Converer, Dataset

```python
class Dataset:

    def __init__(self, dataset_content):
        self.dataset_content = dataset_content
```

```python

class BuilderDataset:

    @staticmethod
    def convert(df ,mapping:Mapping)-> Dataset:
        copy =  df.copy()
        print(mapping.get_columns())
        copy.rename(columns={
            mapping.context : 'context',
            mapping.label:"label",
            mapping.explanation:"explanation",
            mapping.question:"question",
            mapping.question_type:"question_type"
        })

        return copy[mapping.get_columns()]
```
